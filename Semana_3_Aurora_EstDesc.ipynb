## 1. Introduccin a la Inferencia Estadstica

La **inferencia estadstica** es el proceso de utilizar datos de una **muestra** para hacer generalizaciones o inferencias sobre la **poblacin** de la cual se obtuvo la muestra.

En sistemas de control y medicin, la inferencia estadstica es esencial para:

* **Estimar parmetros desconocidos:** Por ejemplo, la cantidad promedio de pedidos recibidos en una cadena logstica o el porcentaje de productos defectuosos en un proceso de fabricacin.
* **Tomar decisiones basadas en datos:** Por ejemplo, determinar si un nuevo sistema de gestin de inventario est mejorando el rendimiento o si se necesitan ajustes en un proceso de produccin para cumplir con los estndares de calidad.

**Conceptos clave:**

* **Poblacin:** Conjunto completo de elementos o individuos que son de inters para un estudio.
* **Muestra:** Subconjunto de la poblacin que se selecciona para el estudio.
* **Parmetro:** Valor numrico que describe una caracterstica de la poblacin (generalmente desconocido).
* **Estimador:** Valor numrico calculado a partir de una muestra que se utiliza para estimar un parmetro poblacional.

## 2. Intervalos de Confianza

Un **intervalo de confianza** es un rango estimado de valores que se obtiene a partir de una *muestra de datos*. Este intervalo se construye con el propsito de indicar, con un *nivel especfico de probabilidad*, dnde se espera que se encuentre el verdadero valor de un **parmetro poblacional desconocido**. Por ejemplo, podramos decir que estamos *95% seguros* de que la estatura promedio de todos los adultos en un pas est entre 170 y 175 cm, basndonos en mediciones de una muestra de 1000 adultos seleccionados al azar. Este rango de 170 a 175 cm sera nuestro **intervalo de confianza**, y el 95% representa nuestro **nivel de confianza** en que el verdadero promedio poblacional se encuentra dentro de ese intervalo.

**Interpretacin del intervalo de confianza:**

La interpretacin correcta de un intervalo de confianza se basa en la repeticin hipottica del proceso de muestreo. Si construyramos mltiples intervalos de confianza del 95% para la estatura promedio de los adultos en un pas, cada uno basado en diferentes muestras de 1000 personas, esperaramos que aproximadamente el 95% de estos intervalos contuvieran el **verdadero valor de la estatura promedio** de la poblacin. Por ejemplo, si realizramos este proceso 100 veces, aproximadamente 95 de los intervalos resultantes incluiran la *estatura promedio real* de todos los adultos del pas. Es importante destacar que cualquier intervalo individual tiene una probabilidad del 95% de contener el verdadero valor, pero no podemos estar seguros de si un intervalo especfico es uno de los que lo contiene o no.

**Nivel de confianza:**

El nivel de confianza, comnmente representado como (1-), es un concepto fundamental en la construccin de intervalos de confianza. Este valor indica la *probabilidad* de que el intervalo calculado contenga el **verdadero valor del parmetro poblacional**. Se expresa tpicamente como un porcentaje, siendo los ms comunes 90%, 95% y 99%. Por ejemplo, en nuestro caso de la estatura promedio, un nivel de confianza del 95% significa que estamos utilizando un mtodo que, a largo plazo, producir intervalos que contienen el verdadero promedio poblacional en el 95% de los casos.

Es importante entender que existe una relacin inversa entre la amplitud del intervalo y el nivel de confianza. Un *mayor nivel de confianza* implica un *intervalo ms amplio*. Por ejemplo, un intervalo de confianza del 99% para la estatura promedio sera ms amplio que uno del 95%, quizs extendindose de 169 a 176 cm en lugar de 170 a 175 cm. Esto refleja el compromiso entre la confianza en nuestras estimaciones y la precisin de las mismas.

**Valor crtico en intervalos de confianza:**

El *valor crtico*, comnmente representado como Z o t, es un componente esencial en la construccin de intervalos de confianza. Este valor est **directamente relacionado con el nivel de confianza** que hemos elegido y la *distribucin de probabilidad* que sigue nuestra estadstica muestral.

En el contexto de nuestro ejemplo sobre la estatura promedio, el valor crtico nos ayuda a determinar *qu tan lejos* debemos extender nuestro intervalo desde la media muestral para alcanzar el nivel de confianza deseado. Por ejemplo, para un nivel de confianza del 95% y asumiendo una distribucin normal, utilizaramos un valor Z de aproximadamente 1.96.

La eleccin entre usar Z o t depende principalmente del *tamao de la muestra* y de si conocemos la **desviacin estndar poblacional**:

- Usamos Z cuando la muestra es grande (generalmente n > 30) o cuando conocemos la desviacin estndar poblacional.
- Optamos por t cuando la muestra es pequea y desconocemos la desviacin estndar poblacional.

En la prctica, estos valores se obtienen de tablas estadsticas o, ms comnmente, mediante funciones en software estadstico. Por ejemplo, si aumentramos nuestro nivel de confianza al 99% para el intervalo de la estatura promedio, nuestro valor crtico Z cambiara a aproximadamente 2.58, resultando en un intervalo ms amplio.

Es crucial entender que el valor crtico *no es arbitrario*, sino que se deriva matemticamente para asegurar que el intervalo resultante tenga exactamente la probabilidad deseada de contener el verdadero parmetro poblacional.

**Pruebas de una cola y dos colas:**

En el contexto de los intervalos de confianza y las pruebas de hiptesis, es importante entender la diferencia entre las pruebas de una cola y dos colas:

1. **Prueba de una cola:** En este tipo de prueba, estamos interesados en si un parmetro poblacional es mayor o menor que un valor especfico, pero no en ambas direcciones simultneamente. ## 3. Pruebas de Hiptesis

En la **prueba de hiptesis**, se utiliza una muestra de datos para evaluar una afirmacin sobre un parmetro poblacional. El proceso se estructura de la siguiente manera:

**Formulacin de hiptesis:**

* **Hiptesis nula (H0):** Es la afirmacin inicial que se prueba. Usualmente establece que no hay efecto o diferencia significativa. Por ejemplo, puede indicar que la media poblacional es igual a un valor especfico.
* **Hiptesis alternativa (H1):** Es la afirmacin que se acepta si se rechaza la hiptesis nula. Esta sugiere la presencia de una diferencia o efecto significativo.

La formulacin de las hiptesis puede indicar si se trata de una prueba de una cola o de dos colas segn cmo se estructuren las hiptesis nula (H0) y alternativa (H1).

1. **Prueba de dos colas:**
   - Se utiliza cuando se quiere detectar cualquier cambio significativo en ambas direcciones (mayor o menor).
   - **Hiptesis nula (H0):** Establece que no hay diferencia o efecto; por ejemplo, $\mu = \mu_0$.
   - **Hiptesis alternativa (H1):** Sugiere que hay una diferencia, sin especificar la direccin; por ejemplo, $\mu \neq \mu_0$.
   - **Indicador clave:** La formulacin de H1 utiliza "" (diferente), lo que implica que se considera cualquier desviacin significativa de $\mu_0$, ya sea hacia arriba o hacia abajo.

2. **Prueba de una cola:**
   - Se utiliza cuando se busca detectar un cambio significativo en una direccin especfica.
   - **Hiptesis nula (H0):** Establece que no hay efecto o que el efecto es en una direccin; por ejemplo, $\mu \leq \mu_0$ o $\mu \geq \mu_0$.
   - **Hiptesis alternativa (H1):** Sugiere un efecto en una direccin especfica; por ejemplo, $\mu > \mu_0$ o $\mu < \mu_0$.
   - **Indicador clave:** La formulacin de H1 incluye una desigualdad direccional ("<" o ">"), lo que implica que se est buscando evidencia de un cambio en una direccin particular.

### Ejemplo:

  **Dos colas:**  
  - **H0:** La media de los tiempos de entrega es igual a 5 das, $\mu = 5$.
  - **H1:** La media de los tiempos de entrega es diferente de 5 das, $\mu \neq 5$.

  **Una cola (superior):**  
  - **H0:** La media de los tiempos de entrega es menor o igual a 5 das, $\mu \leq 5$.
  - **H1:** La media de los tiempos de entrega es mayor a 5 das, $\mu > 5$.

  **Una cola (inferior):**  
  - **H0:** La media de los tiempos de entrega es mayor o igual a 5 das, $\mu \geq 5$.
  - **H1:** La media de los tiempos de entrega es menor a 5 das, $\mu < 5$.

La decisin de utilizar una prueba de una cola o de dos colas depende del contexto de la investigacin y de lo que se espera encontrar. Si solo se espera o se desea detectar una desviacin en una direccin especfica, se utiliza una prueba de una cola. Si se busca detectar cualquier desviacin significativa, se utiliza una prueba de dos colas.

### Pruebas de Hiptesis: Explicacin Detallada y Ejemplos

#### Qu es una prueba de hiptesis?

Una prueba de hiptesis es un procedimiento estadstico utilizado para tomar decisiones o inferencias sobre un parmetro poblacional basndose en una muestra de datos. La prueba evala si la evidencia de la muestra es suficiente para rechazar una hiptesis nula predefinida en favor de una hiptesis alternativa.

#### Contexto Actual: Cadena Logstica de Pedidos

En el contexto de la cadena logstica de pedidos, podramos estar interesados en evaluar si la tasa media de pedidos por hora ha cambiado debido a un nuevo sistema de gestin implementado. Supongamos que antes de la implementacin, la tasa promedio de pedidos por hora era de 10.

#### Planteamiento de la Prueba de Hiptesis

El planteamiento de una prueba de hiptesis sigue estos pasos:

1. **Definicin de las hiptesis**:
   - **Hiptesis nula ($H_0$)**: Es la afirmacin que se busca poner a prueba. Generalmente, se plantea como una afirmacin de igualdad. En este caso, $H_0$: La tasa media de pedidos por hora es 10.
   - **Hiptesis alternativa ($H_1$)**: Es la afirmacin que se considera verdadera si se rechaza la hiptesis nula. Puede ser una afirmacin de diferencia o desigualdad. En este caso, $H_1$: La tasa media de pedidos por hora no es 10.

2. **Seleccin del nivel de significancia ($\alpha$)**: Es la probabilidad de rechazar la hiptesis nula cuando es verdadera. Un valor comnmente utilizado es $\alpha = 0.05$ (5%).

3. **Recoleccin de datos**: Se obtiene una muestra representativa de los datos. En nuestro ejemplo, supongamos que recolectamos datos de 50 horas de pedidos despus de implementar el nuevo sistema.

4. **Clculo del estadstico de prueba**: Utilizando la muestra, se calcula un valor que ayudar a determinar si se rechaza o no la hiptesis nula. En el caso de una prueba de hiptesis para la media, este estadstico puede ser una $z$ o $t$.

5. **Determinacin del valor p**: El valor p es la probabilidad de observar un resultado tan extremo como el de la muestra, bajo la suposicin de que la hiptesis nula es verdadera. Si el valor p es menor que $\alpha$, se rechaza $H_0$.

6. **Decisin**: Basado en el valor p y el nivel de significancia, se decide si se rechaza o no la hiptesis nula.

#### Motivacin para Generar una Prueba de Hiptesis

Una prueba de hiptesis se genera cuando existe una necesidad de tomar decisiones informadas sobre un parmetro poblacional basado en datos muestrales. En el contexto de la cadena logstica, esto podra ser motivado por varios factores:
- **Evaluar cambios**: Verificar si la implementacin de un nuevo sistema de gestin ha alterado significativamente la tasa de pedidos por hora.
- **Validar suposiciones**: Comprobar si las suposiciones operativas actuales (por ejemplo, una tasa de pedidos constante) siguen siendo vlidas.
- **Tomar decisiones estratgicas**: Basar decisiones operativas y estratgicas en evidencia estadstica slida, en lugar de intuiciones o suposiciones no verificadas.

#### Beneficios de Realizar una Prueba de Hiptesis

Realizar una prueba de hiptesis proporciona varios beneficios importantes:
- **Rigor estadstico**: Ofrece un marco formal y riguroso para evaluar afirmaciones sobre un parmetro poblacional.
- **Reduccin de riesgos**: Ayuda a mitigar el riesgo de tomar decisiones basadas en suposiciones incorrectas o en variaciones aleatorias en los datos.
- **Justificacin basada en datos**: Permite justificar decisiones estratgicas y operativas con evidencia cuantitativa, aumentando la credibilidad y confiabilidad de las decisiones tomadas.
- **Identificacin de cambios significativos**: Facilita la deteccin de cambios significativos en los procesos, permitiendo una respuesta oportuna y adecuada a las nuevas condiciones operativas.

Entonces, las pruebas de hiptesis son herramientas fundamentales en el anlisis de datos, permitiendo a las organizaciones tomar decisiones informadas y basadas en evidencia, mejorando as la eficiencia y efectividad de sus operaciones.

El estadstico $z$ se utiliza cuando la muestra es grande (n > 30) y se conoce la desviacin estndar de la poblacin ($\sigma$). El estadstico $z$ se calcula utilizando la siguiente frmula:

$$ z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}} $$

Donde:
- $\bar{x}$ es la media muestral.
- $\mu$ es la media poblacional bajo la hiptesis nula.
- $\sigma$ es la desviacin estndar de la poblacin.
- $n$ es el tamao de la muestra.

El valor $z$ nos indica cuntas desviaciones estndar est la media muestral ($\bar{x}$) de la media poblacional supuesta ($\mu$). Si $|z|$ es grande, esto sugiere que la media muestral est lejos de lo que se esperara si la hiptesis nula fuera verdadera.

#### Qu es el estadstico $t$?

El estadstico $t$ se utiliza cuando la muestra es pequea (n  30) o cuando la desviacin estndar de la poblacin ($\sigma$) es desconocida. En este caso, utilizamos la desviacin estndar muestral ($s$) en lugar de $\sigma$. El estadstico $t$ se calcula utilizando la siguiente frmula:

$$ t = \frac{\bar{x} - \mu}{s / \sqrt{n}} $$

Donde:
- $\bar{x}$ es la media muestral.
- $\mu$ es la media poblacional bajo la hiptesis nula.
- $s$ es la desviacin estndar muestral.
- $n$ es el tamao de la muestra.

El valor $t$ es similar al valor $z$ en su interpretacin, pero toma en cuenta la variabilidad adicional debida al uso de $s$ en lugar de $\sigma$. Por lo tanto, las distribuciones $t$ tienen colas ms gruesas que la distribucin normal, lo que refleja la mayor incertidumbre.

#### Cundo se elige $z$ o $t$?

- Elegir $z$: Se elige el estadstico $z$ cuando la muestra es grande y la desviacin estndar de la poblacin es conocida. En estos casos, la distribucin de la media muestral se aproxima a una distribucin normal.
- Elegir $t$: Se elige el estadstico $t$ cuando la muestra es pequea o la desviacin estndar de la poblacin es desconocida. La distribucin $t$ se ajusta mejor a muestras pequeas y refleja la mayor incertidumbre cuando se estima la desviacin estndar a partir de la muestra.

### Interpretacin:
- Un valor $z$ de 1 significa que la media muestral est a una desviacin estndar de la media poblacional supuesta.
- Un valor $z$ de 2 significa que est a dos desviaciones estndar, y as sucesivamente.
- Cuanto mayor sea $|z|$, ms evidencia tenemos contra la hiptesis nula.

### De $\alpha$ al Valor Crtico

Para comprender cmo se establece la decisin de rechazar o no una hiptesis nula en una prueba de hiptesis, es fundamental entender los conceptos de $\alpha$ (alfa) y el valor crtico.

### $\alpha$ (alfa):
- Es el **nivel de significancia**, que representa la probabilidad de cometer un error Tipo I, es decir, rechazar la hiptesis nula cuando en realidad es verdadera.
- Comnmente, se utiliza  = 0.05 o  = 0.01. Un  ms pequeo indica una menor probabilidad de cometer un error Tipo I, pero tambin puede hacer que la prueba sea menos sensible.
-  determina qu tan estrictos somos al decidir rechazar la hiptesis nula. Un nivel de significancia ms bajo significa que necesitamos evidencia ms fuerte para rechazar la hiptesis nula.

### Valor Crtico:
- Es el punto de corte en la distribucin de la estadstica de prueba que separa la **regin de rechazo** de la **regin de no rechazo**.
- Se determina utilizando  y la distribucin apropiada (z para distribuciones normales y t para distribuciones t de Student).

### Cmo se Obtiene el Valor Crtico:

1. **Para una prueba de dos colas con  = 0.05**:
   - Dividimos  entre 2 porque hay dos colas: 0.05/2 = 0.025.
   - Buscamos en la tabla z el valor que deja un rea de 0.025 en cada cola: z  1.96.

2. **Para una prueba de una cola con  = 0.05**:
   - Buscamos en la tabla z el valor que deja un rea de 0.05 en una cola: z  1.645 (para cola derecha).


```python
# IMPORTANTE: EJECUTA TODO ESTE BLOQUE DE CDIGO DE UNA SOLA VEZ (Shift + Enter en Jupyter)
# Esto asegura que todos los widgets se definan y configuren correctamente.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t
import ipywidgets as widgets
from IPython.display import display, Math, HTML

# --- 1. DECLARACIN DE TODOS LOS WIDGETS CON SUS VALORES INICIALES Y RANGOS POR DEFECTO ---

# Sliders y Dropdowns para Parmetros de Prueba Estadstica
confidence_slider = widgets.FloatSlider(
    value=0.95, min=0.80, max=0.999, step=0.001, description='Confianza:',
    continuous_update=True, orientation='horizontal', readout=True, readout_format='.3f',
)

dist_type_dropdown = widgets.Dropdown(
    options=['Z-distribution (Normal)', 't-distribution'], value='Z-distribution (Normal)',
    description='Distribucin:',
)

sample_size_slider = widgets.IntSlider(
    value=30, min=2, max=100, step=1, description='Tamao de Muestra (n):',
    continuous_update=True, orientation='horizontal', readout=True,
)

test_type_dropdown = widgets.Dropdown(
    options=['Two-tailed', 'One-tailed (Upper)', 'One-tailed (Lower)'], value='Two-tailed',
    description='Tipo de Prueba:',
)

# Inputs para definir los rangos de los sliders REALES
# Estos widgets se declaran y sus valores iniciales son los que se usarn como min/max por defecto.
mean_min_input = widgets.FloatText(
    value=0.0, description='Min Media Real:', readout_format='.1f',
    layout=widgets.Layout(width='auto', flex='1 1 auto')
)
mean_max_input = widgets.FloatText(
    value=200.0, description='Max Media Real:', readout_format='.1f',
    layout=widgets.Layout(width='auto', flex='1 1 auto')
)
std_max_input = widgets.FloatText(
    value=50.0, description='Max Desv. Est. Real:', readout_format='.1f',
    layout=widgets.Layout(width='auto', flex='1 1 auto')
)

# Sliders para valores REALES - Sus rangos se inicializan directamente aqu
# y luego se actualizarn dinmicamente si el usuario cambia los 'min_mean_input' etc.
hypothesized_mean_real_slider = widgets.FloatSlider(
    value=100.0, min=mean_min_input.value, max=mean_max_input.value, step=0.2, description='Media Hipottica ():',
    continuous_update=True, orientation='horizontal', readout=True, readout_format='.1f',
)

sample_mean_real_slider = widgets.FloatSlider(
    value=100.0, min=mean_min_input.value, max=mean_max_input.value, step=0.2, description='Media Muestral (x):',
    continuous_update=True, orientation='horizontal', readout=True, readout_format='.1f',
)

std_dev_real_slider = widgets.FloatSlider(
    value=10.0, min=0.01, max=std_max_input.value, step=0.1, description='Desv. Estndar (/s):',
    continuous_update=True, orientation='horizontal', readout=True, readout_format='.1f',
)

# Output Widgets para la frmula y la regla de decisin
formula_output = widgets.Output()
decision_output = widgets.Output()

# --- 2. DEFINICIN DE FUNCIONES ---

def update_plot(confidence_level, dist_type, sample_size, test_type,
                hypothesized_mean_real, sample_mean_real, std_dev_real):
    # Crear una figura con un solo subplot
    fig, ax = plt.subplots(figsize=(10, 6))

    alpha = 1 - confidence_level
    n = sample_size

    # Asegurarse de que std_dev_real no sea cero para evitar divisin por cero
    if std_dev_real <= 0:
        std_dev_real = 0.001 # Un valor muy pequeo pero no cero

    # Clculo del Error Estndar (utilizando la desviacin estndar en unidades reales)
    if n <= 1:
        standard_error = std_dev_real / 1
    else:
        standard_error = std_dev_real / np.sqrt(n)

    # Clculo del Estadstico de Prueba (Z o t)
    if standard_error == 0:
        test_statistic = 0.0
    else:
        test_statistic = (sample_mean_real - hypothesized_mean_real) / standard_error

    all_relevant_points = [test_statistic]
    dist_char = "Z" # Carcter para la frmula y regla de decisin

    # Inicializar variables para valores crticos y decisiones
    cv1, cv2, cv = None, None, None
    decision_text = ""
    decision_color = 'black'

    # --- Configuracin y Ploteo de la Grfica Principal (Z/t estandarizada) ---
    if dist_type == 'Z-distribution (Normal)':
        x_std = np.linspace(-4, 4, 500)
        y_std = norm.pdf(x_std)
        dist_label = "Distribucin Normal Estndar (Z)"
        dist_char = "Z"

        # Valor crtico para el intervalo de confianza (SIEMPRE de dos colas para el CI)
        ci_critical_value = norm.ppf(1 - alpha / 2)

        if test_type == 'Two-tailed':
            cv1 = norm.ppf(alpha / 2)
            cv2 = norm.ppf(1 - alpha / 2)

            x_accept = x_std[(x_std > cv1) & (x_std < cv2)]
            y_accept = y_std[(x_std > cv1) & (x_std < cv2)]
            ax.fill_between(x_accept, 0, y_accept, color='lightgreen', alpha=0.4, label=f'Regin de Aceptacin ({confidence_level*100:.1f}%)')

            x_reject_left = x_std[x_std <= cv1]
            y_reject_left = y_std[x_std <= cv1]
            ax.fill_between(x_reject_left, 0, y_reject_left, color='red', alpha=0.3, label=f'Regin de Rechazo (/2 = {alpha/2:.3f})')

            x_reject_right = x_std[x_std >= cv2]
            y_reject_right = y_std[x_std >= cv2]
            ax.fill_between(x_reject_right, 0, y_reject_right, color='red', alpha=0.3)

            ax.axvline(cv1, color='blue', linestyle='--', label=f'Valor Crtico 1: {cv1:.3f}')
            ax.axvline(cv2, color='blue', linestyle='--', label=f'Valor Crtico 2: {cv2:.3f}')
            ax.text(cv1 - 0.1, 0.02, f'{cv1:.2f}', rotation=90, verticalalignment='bottom', color='blue')
            ax.text(cv2 - 0.1, 0.02, f'{cv2:.2f}', rotation=90, verticalalignment='bottom', color='blue')
            title_suffix = f"Z = {cv1:.2f}, {cv2:.2f}"

            if test_statistic < cv1 or test_statistic > cv2:
                decision_text = "RECHAZAR HIPTESIS NULA!"
                decision_color = 'red'
            else:
                decision_text = "No rechazar hiptesis nula."
                decision_color = 'darkgreen'

        elif test_type == 'One-tailed (Upper)':
            cv = norm.ppf(1 - alpha)

            x_accept = x_std[x_std < cv]
            y_accept = y_std[x_std < cv]
            ax.fill_between(x_accept, 0, y_accept, color='lightgreen', alpha=0.4, label=f'Regin de Aceptacin ({confidence_level*100:.1f}%)')

            x_reject = x_std[x_std >= cv]
            y_reject = y_std[x_std >= cv]
            ax.fill_between(x_reject, 0, y_reject, color='red', alpha=0.3, label=f'Regin de Rechazo ( = {alpha:.3f})')

            ax.axvline(cv, color='blue', linestyle='--', label=f'Valor Crtico: {cv:.3f}')
            ax.text(cv - 0.1, 0.02, f'{cv:.2f}', rotation=90, verticalalignment='bottom', color='blue')
            title_suffix = f"Z = {cv:.2f}"

            if test_statistic > cv:
                decision_text = "RECHAZAR HIPTESIS NULA!"
                decision_color = 'red'
            else:
                decision_text = "No rechazar hiptesis nula."
                decision_color = 'darkgreen'

        else: # One-tailed (Lower)
            cv = norm.ppf(alpha)

            x_accept = x_std[x_std > cv]
            y_accept = y_std[x_std > cv]
            ax.fill_between(x_accept, 0, y_accept, color='lightgreen', alpha=0.4, label=f'Regin de Aceptacin ({confidence_level*100:.1f}%)')

            x_reject = x_std[x_std <= cv]
            y_reject = y_std[x_std <= cv]
            ax.fill_between(x_reject, 0, y_reject, color='red', alpha=0.3, label=f'Regin de Rechazo ( = {alpha:.3f})')

            ax.axvline(cv, color='blue', linestyle='--', label=f'Valor Crtico: {cv:.3f}')
            ax.text(cv - 0.1, 0.02, f'{cv:.2f}', rotation=90, verticalalignment='bottom', color='blue')
            title_suffix = f"Z = {cv:.2f}"

            if test_statistic < cv:
                decision_text = "RECHAZAR HIPTESIS NULA!"
                decision_color = 'red'
            else:
                decision_text = "No rechazar hiptesis nula."
                decision_color = 'darkgreen'

    else: # t-distribution
        df = max(1, sample_size - 1) # Grados de libertad, mnimo 1
        x_std = np.linspace(-4, 4, 500)
        y_std = t.pdf(x_std, df)
        dist_label = f"Distribucin t de Student (df = {df})"
        dist_char = "t"

        # Valor crtico para el intervalo de confianza (SIEMPRE de dos colas para el CI)
        ci_critical_value = t.ppf(1 - alpha / 2, df)

        if test_type == 'Two-tailed':
            cv1 = t.ppf(alpha / 2, df)
            cv2 = t.ppf(1 - alpha / 2, df)

            x_accept = x_std[(x_std > cv1) & (x_std < cv2)]
            y_accept = y_std[(x_std > cv1) & (x_std < cv2)]
            ax.fill_between(x_accept, 0, y_accept, color='lightgreen', alpha=0.4, label=f'Regin de Aceptacin ({confidence_level*100:.1f}%)')

            x_reject_left = x_std[x_std <= cv1]
            y_reject_left = y_std[x_std <= cv1]
            ax.fill_between(x_reject_left, 0, y_reject_left, color='red', alpha=0.3, label=f'Regin de Rechazo (/2 = {alpha/2:.3f})')

            x_reject_right = x_std[x_std >= cv2]
            y_reject_right = y_std[x_std >= cv2]
            ax.fill_between(x_reject_right, 0, y_reject_right, color='red', alpha=0.3)

            ax.axvline(cv1, color='blue', linestyle='--', label=f'Valor Crtico 1: {cv1:.3f}')
            ax.axvline(cv2, color='blue', linestyle='--', label=f'Valor Crtico 2: {cv2:.3f}')
            ax.text(cv1 - 0.1, 0.02, f'{cv1:.2f}', rotation=90, verticalalignment='bottom', color='blue')
            ax.text(cv2 - 0.1, 0.02, f'{cv2:.2f}', rotation=90, verticalalignment='bottom', color='blue')
            title_suffix = f"t = {cv1:.2f}, {cv2:.2f}"

            if test_statistic < cv1 or test_statistic > cv2:
                decision_text = "RECHAZAR HIPTESIS NULA!"
                decision_color = 'red'
            else:
                decision_text = "No rechazar hiptesis nula."
                decision_color = 'darkgreen'

        elif test_type == 'One-tailed (Upper)':
            cv = t.ppf(1 - alpha, df)

            x_accept = x_std[x_std < cv]
            y_accept = y_std[x_std < cv]
            ax.fill_between(x_accept, 0, y_accept, color='lightgreen', alpha=0.4, label=f'Regin de Aceptacin ({confidence_level*100:.1f}%)')

            x_reject = x_std[x_std >= cv]
            y_reject = y_std[x_std >= cv]
            ax.fill_between(x_reject, 0, y_reject, color='red', alpha=0.3, label=f'Regin de Rechazo ( = {alpha:.3f})')

            ax.axvline(cv, color='blue', linestyle='--', label=f'Valor Crtico: {cv:.3f}')
            ax.text(cv - 0.1, 0.02, f'{cv:.2f}', rotation=90, verticalalignment='bottom', color='blue')
            title_suffix = f"t = {cv:.2f}"

            if test_statistic > cv:
                decision_text = "RECHAZAR HIPTESIS NULA!"
                decision_color = 'red'
            else:
                decision_text = "No rechazar hiptesis nula."
                decision_color = 'darkgreen'

        else: # One-tailed (Lower)
            cv = t.ppf(alpha, df)

            x_accept = x_std[x_std > cv]
            y_accept = y_std[x_std > cv]
            ax.fill_between(x_accept, 0, y_accept, color='lightgreen', alpha=0.4, label=f'Regin de Aceptacin ({confidence_level*100:.1f}%)')

            x_reject = x_std[x_std <= cv]
            y_reject = y_std[x_std <= cv]
            ax.fill_between(x_reject, 0, y_reject, color='red', alpha=0.3, label=f'Regin de Rechazo ( = {alpha:.3f})')

            ax.axvline(cv, color='blue', linestyle='--', label=f'Valor Crtico: {cv:.3f}')
            ax.text(cv - 0.1, 0.02, f'{cv:.2f}', rotation=90, verticalalignment='bottom', color='blue')
            title_suffix = f"t = {cv:.2f}"

            if test_statistic < cv:
                decision_text = "RECHAZAR HIPTESIS NULA!"
                decision_color = 'red'
            else:
                decision_text = "No rechazar hiptesis nula."
                decision_color = 'darkgreen'

    ax.plot(x_std, y_std, color='black', label='Densidad de Probabilidad')

    # Lnea punteada roja para la Media Hipottica (0 estandarizado)
    ax.axvline(0, color='red', linestyle=':', linewidth=2, label=f'Media Hipottica () estandarizada (0)')
    all_relevant_points.append(0) # Asegurar que el 0 se incluya en el rango del eje X

    # Lnea del estadstico de prueba en ax
    ax.axvline(test_statistic, color='purple', linestyle='-', linewidth=2, label=f'Estadstico de Prueba: {test_statistic:.3f}')
    ax.text(test_statistic + 0.05, ax.get_ylim()[1] * 0.5,
             f'{test_statistic:.3f}',
             color='purple', rotation=90, verticalalignment='center', horizontalalignment='left',
             bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='none', alpha=0.7))

    # --- Dibujar el Intervalo de Confianza en la grfica Estandarizada (franja naranja) ---
    # Este CI est centrado en el estadstico de prueba y su ancho es 2 * ci_critical_value
    # Si este CI incluye el 0 (el valor hipotetizado de la media en la escala estandarizada), no se rechaza H0.
    ci_lower_std = test_statistic - ci_critical_value
    ci_upper_std = test_statistic + ci_critical_value

    ci_x_fill = x_std[(x_std >= ci_lower_std) & (x_std <= ci_upper_std)]
    # Usamos la PDF de la distribucin principal para rellenar
    ci_y_fill = norm.pdf(ci_x_fill) if dist_type == 'Z-distribution (Normal)' else t.pdf(ci_x_fill, df)
    ax.fill_between(ci_x_fill, 0, ci_y_fill, color='orange', alpha=0.3,
                    label=f'IC {confidence_level*100:.1f}% para  (centrado en {dist_char})')

    ax.axvline(ci_lower_std, color='darkorange', linestyle=':')
    ax.axvline(ci_upper_std, color='darkorange', linestyle=':')
    ax.text(ci_lower_std, ax.get_ylim()[1]*0.6, f'{ci_lower_std:.2f}', color='darkorange', rotation=90, ha='right', va='center')
    ax.text(ci_upper_std, ax.get_ylim()[1]*0.6, f'{ci_upper_std:.2f}', color='darkorange', rotation=90, ha='left', va='center')


    ax.set_title(f'Valores Crticos y IC para {dist_label}\nNivel de Confianza: {confidence_level*100:.1f}%, : {alpha:.3f} ({test_type})', color='black', fontweight='bold')
    ax.set_xlabel(f'Valor Estandarizado ({dist_type.split("-")[0].strip()} o t)')
    ax.set_ylabel('Densidad de Probabilidad')
    ax.grid(True, linestyle=':', alpha=0.7)
    ax.legend(loc='upper right')
    ax.set_ylim(bottom=0)

    # --- Configurar el Segundo Eje X (Unidades Reales) ---
    # Funcin para transformar de estandarizado a real
    def std_to_real_transform(x_std_val):
        # Esta transformacin asume que el 0 estandarizado corresponde a hypothesized_mean_real
        return hypothesized_mean_real + x_std_val * standard_error

    # Funcin para transformar de real a estandarizado (inversa)
    def real_to_std_transform(x_real_val):
        return (x_real_val - hypothesized_mean_real) / standard_error

    sec_ax = ax.secondary_xaxis('top', functions=(std_to_real_transform, real_to_std_transform))
    sec_ax.set_xlabel(f'Valores en Unidades Reales (escala basada en  = {hypothesized_mean_real:.1f})')

    # Ajustar dinmicamente los lmites del eje X principal para incluir todos los puntos
    all_relevant_points.extend([cv for cv in [cv1,cv2,cv] if cv is not None]) # Aadir todos los CV a los puntos de referencia
    all_relevant_points.extend([ci_lower_std, ci_upper_std]) # Aadir los lmites del CI estandarizado

    if all_relevant_points:
        min_val_std = min(all_relevant_points)
        max_val_std = max(all_relevant_points)

        if max_val_std - min_val_std > 0.01:
            padding_std = (max_val_std - min_val_std) * 0.2
        else:
            padding_std = 1.5

        lower_bound_std = min(min_val_std - padding_std, -4.0)
        upper_bound_std = max(max_val_std + padding_std, 4.0)

        ax.set_xlim(lower_bound_std, upper_bound_std)
    else:
        ax.set_xlim(-4, 4) # Fallback

    plt.tight_layout()
    plt.show()

    # --- Actualizar la frmula con valores reales ---
    formula_output.clear_output(wait=True) # Limpiar la salida anterior
    with formula_output:
        # Numerador de la frmula (diferencia entre media muestral y media hipottica)
        numerator_value_real = sample_mean_real - hypothesized_mean_real

        # Smbolo de la desviacin estndar (sigma para Z, s para t)
        std_dev_symbol = r"\sigma" if dist_type == 'Z-distribution (Normal)' else "s"

        # Construir la parte de la frmula general (ej. (x - ) / ( / n))
        general_denominator_latex = f"{std_dev_symbol} / \\sqrt{{n}}"
        if n <= 1: # Ajustar para el caso n=1 en la visualizacin general de la frmula
             general_denominator_latex = f"{std_dev_symbol}"

        general_formula_part = f"\\frac{{\\bar{{x}} - \\mu_0}}{{{general_denominator_latex}}}"

        # Construir la parte numrica de la frmula (ej. (100-90) / 5.2)
        # Denominador numrico es el error estndar ya calculado.
        numerical_denominator_value_for_display = standard_error if standard_error >= 0.001 else 0.001

        # La parte clave del ajuste: la fraccin numrica es (numerador) / (error_estandar)
        numerical_formula_part = f"\\frac{{{sample_mean_real:.2f} - {hypothesized_mean_real:.2f}}}{{{numerical_denominator_value_for_display:.3f}}}"

        # Componer la frmula completa
        formula_latex = f"$${dist_char} = {general_formula_part} = {numerical_formula_part} = {test_statistic:.3f}$$"
        display(Math(formula_latex))


    # --- Actualizar la regla de decisin y el resultado ---
    decision_output.clear_output(wait=True)
    with decision_output:
        rule_text_ht = ""
        # Regla de decisin de la prueba de hiptesis (en Z/t)
        if test_type == 'Two-tailed':
            if cv1 is not None and cv2 is not None:
                rule_text_ht = f"Rechazar H si {dist_char} < {cv1:.3f} o {dist_char} > {cv2:.3f}"
            else:
                rule_text_ht = "Clculo de valores crticos pendiente."
        elif test_type == 'One-tailed (Upper)':
            if cv is not None:
                rule_text_ht = f"Rechazar H si {dist_char} > {cv:.3f}"
            else:
                rule_text_ht = "Clculo de valores crticos pendiente."
        else: # One-tailed (Lower)
            if cv is not None:
                rule_text_ht = f"Rechazar H si {dist_char} < {cv:.3f}"
            else:
                rule_text_ht = "Clculo de valores crticos pendiente."

        # Regla de decisin del Intervalo de Confianza (equivalente)
        # Aqu, el 0 estandarizado representa la media hipotetizada 
        ci_decision_rule = ""
        if ci_lower_std <= 0 <= ci_upper_std:
            ci_decision_rule = f"0 ( estandarizada) CAE DENTRO del IC [{ci_lower_std:.3f}, {ci_upper_std:.3f}]."
        else:
            ci_decision_rule = f"0 ( estandarizada) NO CAE DENTRO del IC [{ci_lower_std:.3f}, {ci_upper_std:.3f}]."


        # Resumen de la decisin final (consistente con ambas reglas)
        final_decision_html = f"<p style='font-size:1.2em; color: {decision_color};'><b>Decisin Final: {decision_text}</b></p>"

        display(HTML(f"<p style='font-size:1.1em;'><b>Regla de Decisin (Z/t estandarizado):</b> <span style='color:black;'>{rule_text_ht}</span></p>"
                     f"<p style='font-size:1.1em;'><b>Regla de Decisin (Intervalo de Confianza {confidence_level*100:.1f}%):</b> <span style='color:black;'>{ci_decision_rule}</span></p>"
                     f"{final_decision_html}"))

# --- 3. FUNCIONES PARA ACTUALIZAR SLIDERS (REFERENCIAN WIDGETS YA DECLARADOS) ---

# Funcin para actualizar los rangos de los sliders reales
def update_real_ranges(*args):
    # Asegrate de que los valores de min/max sean vlidos (ej. min < max)
    new_min_mean = min(mean_min_input.value, mean_max_input.value - 0.1) # Asegura que min sea menor que max
    new_max_mean = max(mean_max_input.value, new_min_mean + 0.1) # Asegura que max sea mayor que min
    new_max_std = max(std_max_input.value, 0.1) # Desviacin estndar mnima 0.1 para un rango

    # Actualizar sliders de media
    current_val_hyp = hypothesized_mean_real_slider.value
    hypothesized_mean_real_slider.min = new_min_mean
    hypothesized_mean_real_slider.max = new_max_mean
    # Asegurarse de que el valor actual no est fuera del nuevo rango
    hypothesized_mean_real_slider.value = np.clip(current_val_hyp, new_min_mean, new_max_mean)

    current_val_sample = sample_mean_real_slider.value
    sample_mean_real_slider.min = new_min_mean
    sample_mean_real_slider.max = new_max_mean
    # Asegurarse de que el valor actual no est fuera del nuevo rango
    sample_mean_real_slider.value = np.clip(current_val_sample, new_min_mean, new_max_mean)

    # Actualizar slider de desviacin estndar
    current_val_std = std_dev_real_slider.value
    std_dev_real_slider.max = new_max_std
    # Asegurarse de que el valor actual no est fuera del nuevo rango (min ya es 0.01)
    std_dev_real_slider.value = np.clip(current_val_std, 0.01, new_max_std)


# Funcin para actualizar la descripcin del slider de desviacin estndar (sigue siendo relevante para  o s)
def update_std_dev_description(*args):
    if dist_type_dropdown.value == 't-distribution':
        std_dev_real_slider.description = 'Desv. Estndar Muestra (s):'
    else: # Z-distribution
        std_dev_real_slider.description = 'Desv. Estndar Poblacin ():'

# --- 4. CONFIGURAR OBSERVADORES Y LA INTERFAZ ---

# Observar cambios en los inputs de rango
# 'continuous_update=False' para que no se actualice con cada tecleo, solo al finalizar la edicin
mean_min_input.observe(update_real_ranges, 'value')
mean_max_input.observe(update_real_ranges, 'value')
std_max_input.observe(update_real_ranges, 'value')

# Observar cambios en el dropdown de tipo de distribucin
dist_type_dropdown.observe(update_std_dev_description, 'value')

# Enlazar la funcin de actualizacin del grfico con los widgets
plot_interactive_output = widgets.interactive_output(update_plot, {
    'confidence_level': confidence_slider,
    'dist_type': dist_type_dropdown,
    'sample_size': sample_size_slider,
    'test_type': test_type_dropdown,
    'hypothesized_mean_real': hypothesized_mean_real_slider,
    'sample_mean_real': sample_mean_real_slider,
    'std_dev_real': std_dev_real_slider
})

# Panel de control con los widgets
controls_panel = widgets.VBox([
    widgets.HTML("<h3>Ajustar Rangos de Valores Reales:</h3>"),
    widgets.HBox([widgets.Label('Rango de Medias:'), mean_min_input, mean_max_input]),
    widgets.HBox([widgets.Label('Max Desv. Est. Real:'), std_max_input]), # Aadir etiqueta para este input
    widgets.HTML("<hr><h3>Parmetros de Prueba Estadstica:</h3>"),
    widgets.HBox([widgets.Label('Nivel de Confianza:'), confidence_slider]),
    widgets.HBox([widgets.Label('Distribucin:'), dist_type_dropdown]),
    widgets.HBox([widgets.Label('Tipo de Prueba:'), test_type_dropdown]),
    widgets.HBox([widgets.Label('Tamao de Muestra (n):'), sample_size_slider]),
    widgets.HTML("<hr><b>Simulacin del Estadstico de Prueba (con valores reales)</b><hr>"),
    formula_output, # Aqu se muestra la frmula
    widgets.HBox([widgets.Label('Media Hipottica ():'), hypothesized_mean_real_slider]),
    widgets.HBox([widgets.Label('Media Muestral (x):'), sample_mean_real_slider]),
    widgets.HBox([widgets.Label('Desv. Estndar (/s):'), std_dev_real_slider]),
])

# Mostrar el panel de control, el grfico interactivo y la salida de decisin
display(widgets.VBox([controls_panel, plot_interactive_output, decision_output]))

# --- 5. LLAMADAS INICIALES PARA SINCRONIZAR ESTADO (AHORA MS SEGURAS) ---
# Estas llamadas se hacen despus de que todos los widgets han sido instanciados.
update_std_dev_description() # Configura la descripcin inicial de la desviacin estndar
update_real_ranges() # Configura los rangos iniciales de los sliders reales
```
